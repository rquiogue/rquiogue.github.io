In the future we know that there are many things that go into where we choose where we want to live. In this test we want to see if there is any correlation between the two

```python
coef, p_val = stats.pearsonr(data2019['Healthy life expectancy'], data2019['Freedom to make life choices'])

print(f"Pearson Correlation Coefficient: {coef:.2f}")
print(f"P-value: {p_val}")
```

Using the pearson correlation coefficent (add link) we are able to measure the relationship between two variables, this measurement ranges from -1 ( negative correlation) to 1 (positive correlation) with a coefficent of close to 0 meaning that there is NO relationship.

In our scenario we recvied a coefficient of 0.39 which is close to 0 but also a good amount away for us to consider its correlation in the long run.

The p value will tell us the signiciance of the correlation coefficiet and will tell us if the null hypothesis is true. Typically a small p value tells us that the corrleation coefficient is true and in this case we recivied a small p value

Using this information we can make a graph to better visualize the correlation.

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='Healthy life expectancy', y='Freedom to make life choices', data=data2019, color='green')
plt.title('Plot of Healthy vs Freedom')
plt.xlabel('Healthy')
plt.ylabel('Freedom')
plt.text(data2019['Healthy life expectancy'].min(), data2019['Freedom to make life choices'].max(),
         f'Pearson Correlation Coefficient: {coef:.2f}', fontsize=12, color='red')
plt.plot(data2019['Healthy life expectancy'], coef * data2019['Healthy life expectancy'], color='blue', linestyle='--')
plt.show()
```

Based on the graph and calculations provided, we observe a correlation, although it is not significant. The calculated correlation coefficient is 0.39, indicating a positive correlation, but it's not substantial enough for us to conclude that it is a strong correlation.

There are are some outliers that have a very healthy life expectancy but very little freedom and the opposite is true with some having higher freedom but lower healthy life expectancy.

While there is not a huge correlation between the two I think it is important to keep this in mind because the correlation is in fact in the positive direction. So in the end we believe this is something that this relationship between freedom and health is essential to consider

**Test 3**


We want to test to see if the average happiness score of the world is greater than 5. We are testing to see if the world is generally happier than it is unhappy

---

H0: μ = 5

HA: μ > 5

α = 0.05

---

```python
happinessScores = data2019.Score
mu = 5
alpha = 0.05

print("The mean happiness score: " + str(happinessScores.mean()))
print("n = " + str(len(happinessScores)))

t_stat, p_value = stats.ttest_1samp(happinessScores, mu, alternative='greater')
print("t_stat = " + str(t_stat))
print("p-value = " + str(p_value))

plt.figure(figsize=(10, 6))
sns.histplot(happinessScores, bins=20, color='skyblue')
plt.axvline(x=happinessScores.mean(), color='red', linestyle='--', label='Mean')
plt.axvline(mu, color='green', linestyle='--', label='Mu')
plt.xlabel('Happiness Score')
plt.ylabel('Frequency')
plt.title('Distribution of Happiness Scores')
plt.show()

if p_value < alpha:
    print("We reject the null hypothesis. We have enough evidence to say that mean is greater than 5")
else:
    print("We fail to reject the null hypothesis. We do not have the evidence to say that the mean is greater than 5")
```

Based on the hypothesis test we can see that the average happiness scores are in fact greater than 5. This means that the on average the world is happier than unhappier. The p-value was very low so there is a very low probabability that we falsely rejected the null

# **Primary Analysis**

In this section, the technology of machine learning will be used in order to further explore this data and the specific characteristics of the set. Machine learning is a subfield of artificial intelligence (AI) that focuses on the development of algorithms and statistical models that enable computer systems to learn and improve their performance on a specific task or problem without being explicitly programmed. Due to the restrictions of our data set, we will be using unsupervised learning in order to find clusterings of data. Unsupervised learning is a subset of machine learning in which the input data does not have any labels. Instead, it finds specific groupings or "Clusters"

In this we will specifially using K-means clustering. It uses clusters and centroids in order to find groupings. A cluster is a set of instances with similar characteristics. A "centroid" is a point that represents the center of each cluster

```python
clustering_df = data2019.filter(['Healthy life expectancy', 'Social support'], axis = 1)
p = sns.scatterplot(data=data2019, x='Healthy life expectancy', y='Social support')
p.set_xlabel('Healthy life expectancy', fontsize=14)
p.set_ylabel('Social support', fontsize=14)
```

Here in this first section of code we are starting to visualize what our data looks like and maybe understand what kind of clusters there are.

```python
WCSS = []
k = [1, 2, 3, 4, 5]
for j in k:
    kmModel = KMeans(n_clusters=j)
    kmModel = kmModel.fit(clustering_df)
    WCSS.append(kmModel.inertia_)

ax = plt.figure().gca()
plt.plot(k, WCSS, '*-')
plt.xlabel('Number of clusters (k)', fontsize=14)
plt.ylabel('Within-cluster sum of squares (WCSS)', fontsize=14)
```

This graph shows the Within-Cluster sum of squares for each given "K". K in the K means clustering algorithm is a hyperparameter that changes the amount of clusters and centroids shown. Here we use the "elbow" method in order to find the optimal number of centroids. This method uses the graph shown and finds where the

With this analysis we are looking into the relationship between healthy life expectancy and social support and see if we can find any groups that they have within them

```python
kmModel = KMeans(n_clusters = 2)

kmModel = kmModel.fit(clustering_df)

centroids = kmModel.cluster_centers_
```

Here we are setting up the model to fit the data set. This is where we train and obtain the centroids.

```python
clusters = kmModel.fit_predict(clustering_df[['Healthy life expectancy', 'Social support']])

p = sns.scatterplot(
    data=clustering_df, x='Healthy life expectancy', y='Social support', hue=clusters, style=clusters
)
p.set_xlabel('Healthy life expectancy', fontsize=14)
p.set_ylabel('Social support', fontsize=14)

# Add centroid for cluster 0
plt.scatter(x=centroids[0, 0], y=centroids[0, 1], c='black')

# Add centroid for cluster 1
plt.scatter(x=centroids[1, 0], y=centroids[1, 1], c='black', marker='X')
```

Here we have two separate groups shown in orange and blue. We have each centroid also shown. With this we gain insight on the different groupings of the data set with it looking like it is split between the top right quadrant and then the bottom left quadrant. These two factors: Healthy life expectancy and social support are very important and we ideally want a country that can balance both. With these clusterings we can see that there is a clear distinction between the countries with high life expectancy and social support and the ones without.

We will now be doing something similar but with 3 variables instead of two

```python
clustered_df_3d = data2019.filter(['GDP per capita', 'Freedom to make life choices', 'Generosity'], axis = 1)
WCSS = []
k = [1, 2, 3, 4, 5]
for j in k:
    kmModel = KMeans(n_clusters=j)
    kmModel = kmModel.fit(clustered_df_3d)
    WCSS.append(kmModel.inertia_)

ax = plt.figure().gca()
plt.plot(k, WCSS, '*-')
plt.xlabel('Number of clusters (k)', fontsize=14)
plt.ylabel('Within-cluster sum of squares (WCSS)', fontsize=14)
```

Here we have shown that the optimal amount of clusters is two so we will be using a kmeans with k being set to 2

```python
kmModel_3d = KMeans(n_clusters = 2)
kmeans_3d = kmModel_3d.fit_predict(clustering_df)

centroids = kmModel_3d.cluster_centers_
```

From there we set up the model and obtain the clusters and centroids

```python
clustered_df_3d['cluster'] = kmeans_3d

cluster1 = clustered_df_3d[clustered_df_3d.cluster==0]
cluster2 = clustered_df_3d[clustered_df_3d.cluster == 1]
```

Here we split up the clusters into two separate data frames so we can plot them

```python
kplot = plt.axes(projection='3d')

kplot.scatter(cluster1['GDP per capita'], cluster1['Freedom to make life choices'], cluster1['Generosity'], c='red', label = 'Cluster 1')
kplot.scatter(cluster2['GDP per capita'], cluster2['Freedom to make life choices'], cluster2['Generosity'], c='green', label = 'Cluster 1')

plt.scatter(kmModel_3d.cluster_centers_[:,0], kmModel_3d.cluster_centers_[:,1], color = 'indigo', s = 200)
plt.legend()
plt.title("Kmeans")
plt.show()

```

Here is our plot from here we can gain insights on how to model has clustered data points. Here we have two distinct groups each shown in red and green respectively

With both of these models we can figure out the different groups that our prospective country might belong to. This can help us narrow down the search as looking at all of the data at once can be challenging. We can narrow our search based on being in a favorable cluster.

Using these models we are able to take in specific characteristics that are favorable to a person and help them narrow down their search of where they would want to live. This answers our question of how to determine where to live !
